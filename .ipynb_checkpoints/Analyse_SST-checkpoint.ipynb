# Analyse du fonctionnement d'une sous-station

Maintenant que la modélisation d'une sous-station n'a plus de secret pour vous, nous vous proposons d'analyser le fonctionnement de la sous-station modélisée. Pour cela, le code que vous avez élaboré au précédent notebook a été converti en deux fonctions distinctes :
- une fonction nommée **"SST_nom"** qui permet de déterminer les <ins>caractéristiques nominales</ins> de la sous-station lors de sa création comme objet Python
- une fonction nommée **"SST_fonct"** qui permet de calculer un point de fonctionnement de la sous-station en <ins>conditions non-nominales</ins>:

Ces fonctions sont écrites dans le fichier `ComposantsRDC.py` présent dans le dossier `\\Modèles`.
Vous découvrirez ci-dessous la nouvelle synthaxe du code.

**Importation des bibliothèques**

```python
import os
import numpy as np
import pandas as pd
from math import exp
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib widget

dossier = os.getcwd()
os.chdir(dossier+'\\Modèles')
import ComposantsRDC as RDC
os.chdir(dossier)
```

**Données d'entrées et constantes**

```python
#Chargement des données d'entrée
os.chdir(dossier+'\\Entrées')
Input = pd.read_excel('Entree_SST_Test.xlsx') #lecture des données d'entrées dans le fichier Excel
temps_simu = len(Input) #nombre de pas de temps
os.chdir(dossier)

#Constantes physiques
Cp=4180
Rho=1000
```

**Définition du point nominal**

Rappel: la sous-station a une puissance nominale de 20 kW, une efficacité de 0,9. Elle alimente un réseau secondaire fonctionnant sur un régime nominal de température 80/60°C. Le régulateur a un gain de 0,002

```python
#Caractéristiques nominales connues de la SST
Kp=0.002
P_nom =20000
E_nom =0.9
Ts_in_nom =60
Ts_out_nom =80

#Détermination du point nominal
Carac_nom=RDC.SST_nom(Kp,Ts_out_nom,Ts_in_nom,E_nom,P_nom)
```

**Initialisation du calcul**

```python
Tp_in = 65*np.ones(temps_simu)

Text = np.array(Input['Text']) # Vecteur des températures extérieures
P_SST = np.array(Input['P (W)']) # Vecteur des puissances appelées côté secondaire

SST_State = np.zeros((temps_simu,8)) #Vecteur des sorties calculées (P_fournie, Tcons, Ts_out, Tp_out, Mp, Ts_in, inconfort, Youv)
SST_State[0,7] = 0.5  #Initialisation de l'ouverture de la vanne
```

**Boucle de calcul**

```python
for t in range(1,temps_simu):
    SST_State[t]=RDC.SST_fonct(Carac_nom,SST_State[t-1,7],Tp_in[t], P_SST[t],Text[t])
```

**Sortie graphique**

```python
plt.clf()

ax_t = np.arange(1,temps_simu)
plt.plot(ax_t, SST_State[1:,2], label = 'Ts_out')
plt.plot(ax_t, SST_State[1:,1], label = 'Tcons')
plt.legend()
plt.xlabel='Temps (10 min)'
plt.ylabel='Température (°C)'
plt.grid(True)
plt.show()
```

## Configuration de référence

On étudie une sous-station de puissance nominale de 120 kW, d'efficacité 0.9 qui alimente un réseau secondaire fonctionnant sur le régime 80/60°C. Le gain du régulateur est fixé à 0,002

```python
#Caractéristiques nominales connues de la SST
Kp=0.002
P_nom =120000
E_nom =0.9
Ts_in_nom =60
Ts_out_nom =80
```

```python
#Détermination du point nominal
Carac_nom=RDC.SST_nom(Kp,Ts_out_nom,Ts_in_nom,E_nom,P_nom)

#Initialisation
Tp_in = 65*np.ones(temps_simu)
Text = np.array(Input['Text']) # Vecteur des températures extérieures
P_SST = np.array(Input['P (W)']) # Vecteur des puissances appelées côté secondaire
SST_State = np.zeros((temps_simu,8)) #Vecteur des sorties calculées (P_fournie, Tcons, Ts_out, Tp_out, Mp, Ts_in, inconfort, Youv)
SST_State[0,7] = 0.5  #Initialisation de l'ouverture de la vanne

#Boucle de calcul
for t in range(1,temps_simu):
    SST_State[t]=RDC.SST_fonct(Carac_nom,SST_State[t-1,7],Tp_in[t], P_SST[t],Text[t])
SST_reference=SST_State

#Sorties graphiques
fig1, axes = plt.subplots(nrows=2, ncols=2, figsize=(9,10))
#fig1.suptitle("""Cas (c) : efficacité de 0,7""")
ax0, ax1, ax2, ax3 = axes.flatten()

ax0.plot(np.arange(0,temps_simu),SST_reference[:,1], label = "T° de consigne au secondaire")
ax0.plot(np.arange(0,temps_simu),SST_reference[:,2], label = "T° de départ au secondaire")
ax0.set_title('Température côté secondaire')
ax0.legend()
ax0.grid()
ax0.set_xlabel('Temps (pas de 10 min)')
ax0.set_ylabel('Température (°C)')

ax1.plot(np.arange(0,temps_simu),SST_reference[:,3], label = "T° sortie côté primaire")
ax1.plot(np.arange(0,temps_simu),Tp_in, label = "T° entrée côté primaire")
ax1.set_title('Températures côté primaire')
ax1.legend()
ax1.grid()
ax1.set_xlabel('Temps (pas de 10 min)')
ax1.set_ylabel('Température (°C)')

ax2.plot(np.arange(0,temps_simu),SST_reference[:,7]*100)
ax2.set_title('Ouverture de la vanne côté primaire')
ax2.set_xlabel('Temps (pas de 10 minutes)')
ax2.set_ylabel('Ouverture (%)')
#ax2.legend()
ax2.grid()

ax3.plot(np.arange(0,temps_simu),SST_reference[:,4])
ax3.set_title('Débit dans la sous-station côté primaire')
ax3.set_xlabel('Temps (pas de 10 minutes)')
ax3.set_ylabel('Débit (kg/s')
#ax3.legend()
ax3.grid()

os.chdir(dossier+'\\Résultats')
fig1.savefig("reference.png")
os.chdir(dossier)

fig1.tight_layout()
fig1.canvas.toolbar_position = 'right'
    
```
