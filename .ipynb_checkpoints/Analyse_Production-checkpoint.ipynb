# Analyse du modle de centrale de production

Dans ce notebook, est prsent la modlisation de la centrale de production du rseau de chaleur.
La centrale de production dans notre cas est compos de :
- Un changeur gothermie
- une chaudire gaz
- Une pompe
- Un systme de rgulation de la temprature de dpart


<div>

</div>
<center> Schma d'une centrale de production de chaleur </center>










**Modlisation de lchangeur gothermal**

Lchangeur de chaleur gothermal est un changeur  plaque mont en contre-courant. La temprature de la source gothermale T_geo_in est considre constante toute au long de lanne et gale  65C. Les contraintes rglementaires sur le dbit de puisage maximal M_geo dans la nappe et sur la temprature de rejet minimale T_geo_out imposent que le dbit de puisage dans la nappe est toujours infrieur au dbit total du rseau (M_geo< M_res)).   Dans un souci de simplicit, on choisit de modliser lchangeur par un modle  efficacit constante E=0,95.

<div>

</div>
<center> Schma d'une centrale de production de chaleur </center>


**Modlisation de la chaudire gaz**

La chaudire gaz et lchangeur gothermal sont monts en srie. De ce fait, la chaudire gaz vient en appoint de lchangeur gothermal afin datteindre la temprature de consigne de dpart du rseau si la temprature en sortie de lchangeur gothermal est trop basse. On considrera que le rendement de la chaudire  est gal  0,85.

**Rgulation de la temprature de dpart au primaire**

La temprature du rseau en sortie du site de production est rgule via une loi deau linaire. Ainsi, la temprature de dpart du rseau volue en fonction de la temprature extrieure, plus la temprature extrieure est basse plus la temprature de dpart doit tre leve pour fournir la temprature ncessaire aux habitations. En outre, les rseaux ne fournissent pas uniquement du chauffage mais aussi de leau chaude sanitaire (pas modlise dans notre cas). Ainsi pour des raisons sanitaires (lgionellose) la temprature de lECS ne peut pas tre infrieure  60C, donc la temprature de dpart du rseau ne peut pas tre infrieure  60C. La loi deau doit tre dimensionne en prenant en compte les clients qui demandent la temprature la plus leve et les pertes dans les canalisations. 

La loi d'eau au dpart du rseau primaire comporte donc deux parties :
- une partie affine o la temprature de dpart diminue lorsque la temprature extrieure augmente,
- une partie constante  60C pour des tempratures suprieures  10C environ dans notre exemple.

<div>

</div>
<center> Schma de la loi d'eau au dpart du circuit primaire </center>

Dans notre exemple, la partie affine a pour ordonne  l'origine 77 C et comme pente -1,85 C/C, ce qui donne une temprature de dpart de 90C pour une temprature d'air extrieur de -7C.

# Prsentation du code

```python
import numpy as np
import math
import os
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib widget

dossier = os.getcwd()
```

**Paramtres**

```python
temps_simu = 72 #temps en heures
pas_de_temps = 10 # pas de temps des entres et sorties en minutes
nb_pdt = int(temps_simu*60/pas_de_temps)
delta_t = 1 # pas de temps de calcul en minutes
nb_dt = int(pas_de_temps/delta_t)
```

**Constantes**

```python
Rho = 1000 # kg/m3
Cp = 4180 # J/(kg.K)
```

**Caractristiques de la centrale de production**

```python
E_geo = 0.95
eta_gaz = 0.85
T_geo_base = 65
```

**Entres de la simulation**

```python
"Dbit en entre"
M_res_nom = 10  # kg/s       "Dbit nominal

M_res = np.ones(nb_pdt)*M_res_nom*0.8

"Temprature extrieure "
#Evolution sinusodale de la temprature
Text = [5+5*math.sin(math.pi*t*10/(12*60)-math.pi/2) for t in range(nb_pdt)]

"Temprature de l'eau en retour du rseau primaire (entre production)"
#Evolution sinusodale de la temprature
T_res_ret = [50+10*math.sin(math.pi*t*10/(12*60)-math.pi/2*1.2) for t in range(nb_pdt)]
```

**Initialisation**

```python
T_res_dep= np.zeros(nb_pdt) # Temprature dpart primaire
T_cons_prod = np.zeros(nb_pdt)
P_geo = np.zeros(nb_pdt)
P_gaz = np.zeros(nb_pdt)
P_gaz_reel = np.zeros(nb_pdt)
P_res_tot = np.zeros(nb_pdt)
T_cons_prod = np.zeros(nb_pdt)
T_geo_out = np.zeros(nb_pdt)
```

**Calcul de la consommation de gaz et nergie fournie**

```python
for t in range(nb_pdt-1):
    "Calcul de la temprature de consigne au dpart primaire"
    T_cons_prod1 = -1.85 * Text[t] + 77 #loi de chauffe primaire
    if T_cons_prod1 < 60:
            T_cons_prod1 = 60
    else:
            T_cons_prod1 = -1.85 * Text[t] +77
    T_cons_prod[t] = T_cons_prod1
    
    "calcul de la temprature de sortie de l'changeur gothermal"
    if T_res_ret[t] > T_cons_prod1:
        T_res_dep = T_res_ret
        P_geo[t] = 0
        P_gaz[t]= 0       
    else:
        if T_res_ret[t] > T_geo_base:
            T_geo_out[t] = T_res_ret[t]
        else:
            T_geo_out[t] = E_geo * ( T_geo_base - T_res_ret[t] ) +T_res_ret[t] # Calcul de la temprature de sortie rseau de l'echangeur geothermie
    
        if T_geo_out[t] > T_cons_prod1:
            T_geo_out[t]= T_cons_prod1
            P_geo[t] = Cp * M_res[t] * (T_geo_out[t] - T_res_ret[t] ) # Calcul puissance geothermie
        else:
             P_geo[t] = Cp * M_res[t] * (T_geo_out[t] - T_res_ret[t]) # Calcul puissance geothermie
    
        "calcul de la puissance d'appoint gaz"
        P_gaz[t] = M_res[t] * Cp * (T_cons_prod[t] - T_geo_out[t]) # Calcul puissance chaudire gaz
       
    P_gaz_reel[t] = P_gaz[t] / eta_gaz # calcul puissance gaz relle
    
    "puissance totale fournie au rseau"
    P_res_tot[t] = P_geo[t] + P_gaz[t]  # calcul puissance totale fournie au rseau
    
#sorties : P_geo, P_gaz_reel, P_res_tot, T_cons_prod, T_geo_out
print('\033[1m'+ f"Le taux d'nergie renouvelable est de {np.round(100*sum(P_geo)/sum(P_res_tot),2)}%")
```

**Sorties graphiques**

```python
ax_t = np.arange(0,temps_simu,pas_de_temps/60)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,6))
ax0, ax1, ax2, ax3 = axes.flatten()

ax0.plot(ax_t, Text,label = 'T ext')
ax0.set_title('Temprature extrieure (C)')
ax0.set(xlabel='Temps (h)', ylabel='Temprature extrieure (C)')
ax0.grid()

ax1.plot(ax_t, T_res_ret, label = 'T retour rseau')
ax1.plot(ax_t, T_cons_prod, label = 'T de consigne')
ax1.set_title('Tempratures')
ax1.legend()
ax1.set(xlabel='Temps (h)', ylabel='Temprature (C)')
ax1.grid()

Texterieur=np.linspace(-7, 20)
Tconsigne = -1.85 * Texterieur + 77
ax2.plot(Texterieur, Tconsigne)
ax2.set_title('Loi de chauffe')
ax2.grid()

ax3.plot(ax_t, P_geo,label = 'Gothermie')
ax3.plot(ax_t, P_gaz,label = 'Gaz')
ax3.legend()
ax3.set_title('Production de chaleur')
ax3.grid()

fig.tight_layout()
fig.canvas.toolbar_position = 'right'

os.chdir(dossier+'\\Rsultats')
fig.savefig("Test_production.png")
os.chdir(dossier)

plt.show()
```
